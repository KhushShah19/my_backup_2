```python
def name_in_dicts(full_list=list):
    new_list = []
    try:
        full_list = ast.literal_eval(full_list)
        for each_dict in full_list:
            new_list.append(each_dict['name'])
    except:
        new_list = []
    return new_list
```

```python
def save_json(data, file_name): # C:\VS_Code_folder\Mongodb\movie_review\data
    file_path = 'C:/VS_Code_folder/Mongodb/movie_review/data/' + file_name + ".json"
    formatted_json = json.dumps(data, indent=4, sort_keys=True)
    with open(file_path, 'w') as file:
        file.write(formatted_json) 
```
